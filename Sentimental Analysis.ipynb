```python
import pandas as pd # for reading and writing csv files
import numpy as np  #to perfrom numerical operations
import re # text cleaning

from sklearn.model_selection import train_test_split #for splitting data into training and testing
from sklearn.feature_extraction.text import TfidfVectorizer # for converting text into nuerical features
from sklearn.linear_model import LogisticRegression  #to build regresssion model
from sklearn.metrics import accuracy_score, classification_report # to check the accuracy of the model
```

```python
df= pd.read_csv(r"C:\Users\Ausaaf Siddiqui\OneDrive\Desktop\Sentimental Analysis\imdb_top_1000.csv")
```

```python
df.head(10)
```

```python
def clean_text(text):
    text=text.lower() #convert the text into lowercase
    text= re.sub(r'[^a-zA-Z]', ' ', text) #remove every single punctuations, numbers nd symbols
    return text
```

```python
df['cleaned'] = df['Overview'].apply(clean_text)
```

```python
vectorizer=TfidfVectorizer(max_features=5000) # used only for 5000 important words
X= vectorizer.fit_transform(df['cleaned']).toarray() #covert all the text to numbers
Y= df['IMDB_Rating']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Check what your ratings look like
print(y_train.describe())
#  This shows min, max, average of your ratings

# Check what happens after mapping
mapped = y_train.map(lambda x: 1 if x >= 7.0 else 0)
print(mapped.value_counts())
# If you see only "1" and no "0" → confirmed!
```

```python
# If ratings are between 7-10, try 8.0 as cutoff
model.fit(X_train, y_train.map(lambda x: 1 if x >= 9.0 else 0))

# Or split at middle of your range
model.fit(X_train, y_train.map(lambda x: 1 if x >= 8.0 else 0))
```

```python
y_pred= model.predict(X_test) #test on unseen data

print("Accuracy: ", accuracy_score(y_test.map(lambda x: 1 if x >= 9.0 else 0), y_pred))
print(classification_report(y_test.map(lambda x: 1 if x >= 8.0 else 0), y_pred))
```

```python
import streamlit as st
import pickle

# Load saved model and vectorizer
model = pickle.load(open("sentiment_model.pkl", "rb"))
vectorizer = pickle.load(open("vectorizer.pkl", "rb"))

def predict(text):
    vec = vectorizer.transform([text]).toarray()  # Fix: [text] not ()
    return model.predict(vec)[0]

st.title("Sentiment Analysis App")           # App heading

user_input = st.text_input("Enter text")     # Text box for user

if st.button("Predict"):                     # Button
    result = predict(user_input)
    st.write("Sentiment:", result)           # Show result
```
